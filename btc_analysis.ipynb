```python
btc_prices = pd.read_csv("btc_price.csv", header=0, index_col=0, parse_dates=True)
spx_prices = pd.read_csv("spx_prices.csv", header=0, index_col=0, parse_dates=True)
```

```python
btc_prices.head()
```

```python
btc_prices = btc_prices.iloc[::-1]
spx_prices = spx_prices.iloc[::-1]
spx_prices.head()
btc_prices.head()
```

```python
spx_close = spx_prices['Close/Last']
```

```python
spx_close.head()
```

```python
btc_close = btc_prices['Close/Last']
btc_close.head()
```

```python
btc_returns = btc_close.pct_change()
spx_returns = spx_close.pct_change()
```

```python
spx_returns = spx_returns.dropna()
```

```python
btc_returns = btc_returns.dropna()
```

```python
spx_returns.head(), btc_returns.head()
```

```python
returns = pd.concat([btc_returns, spx_returns], axis=1)
```

```python
returns.columns = ['BTC', 'SPX']
```

```python
returns
```

```python
returns = returns.dropna()
returns.plot.line()
```

```python
returns.std()
```

```python
import numpy as np
```

```python
returns.std() * np.sqrt(12)
```

```python
returns.shape
```

```python
total_return = (1+returns).prod() - 1
```

```python
total_return * 100
```

```python
return_per_day = (1+returns).prod() ** (1/returns.shape[0]) - 1
```

```python
return_per_day
```

```python
annualized_return = (1+return_per_day)**252 - 1
```

```python
annualized_return
```

```python
annualized_vol = returns.std() * np.sqrt(12)
```

```python
annualized_vol
```

```python
annualized_return / annualized_vol
```
