```python
btc_returns = algo.ReturnSeries("data/btc_monthly_5years.csv", "BTC", "Close", False)
```

```python
returns = compute_returns(["data/btc_monthly_5years.csv", 
                           "data/spx_monthly_5years.csv"],
                          ["Close", "Close"],
                          ["BTC", "SPX"])
```

```python
def drawdown(principal, returns):
    dds = []
    for col in returns.columns:
        wealth_index = principal * (1+returns[col]).cumprod()
        previous_peaks = wealth_index.cummax()
        dd = (wealth_index - previous_peaks)/previous_peaks
        dds.append(dd)
    return dds
```

```python
drawdowns = drawdown(1000, returns)
for dd in drawdowns:
    dd.plot()
```
